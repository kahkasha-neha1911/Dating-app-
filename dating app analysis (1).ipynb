# Standardize Categorical Data

```python
# Standardizing categorical fields
categorical_columns = ['diet', 'body_type', 'education', 'drugs', 'smokes']

# Convert to lowercase and strip spaces
df[categorical_columns] = df[categorical_columns].apply(lambda x: x.str.lower().str.strip())

# Replace inconsistencies manually if needed
df['diet'].replace({'strictly any': 'strict', 'mostly oth': 'mostly other'}, inplace=True)
df['body_type'].replace({'a little extr': 'a little extra'}, inplace=True)
df['education'].replace({'working on': 'in progress', 'graduated from masters program': 'master’s degree'}, inplace=True)
df['drugs'].replace({'never': 'no', 'sometimes': 'occasional'}, inplace=True)
df['smokes'].replace({'sometimes': 'occasional', 'no': 'never'}, inplace=True)
```

# Fix Encoding Issues in Text Fields

```python
# List of essay columns
essay_cols = ['essay0', 'essay1', 'essay2', 'essay3', 'essay4', 'essay5', 'essay6', 'essay7', 'essay8', 'essay9']

# Fix encoding issues and remove HTML tags
import html

for col in essay_cols:
    df[col] = df[col].astype(str)  # Convert to string
    df[col] = df[col].apply(lambda x: html.unescape(x))  # Fix encoding issues
    df[col] = df[col].str.replace(r'<br\s*/?>', ' ', regex=True)  # Remove HTML tags
```

# Separate Multiple Categories

```python
# Split ethnicity into multiple binary columns
ethnicities = df['ethnicity'].str.get_dummies(sep=', ')
df = pd.concat([df, ethnicities], axis=1)

# Drop the original ethnicity column
df.drop(columns=['ethnicity'], inplace=True)
```

# Convert Numerical Fields to Proper Formats

```python

# Convert to numeric, forcing errors to NaN if necessary
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['income'] = pd.to_numeric(df['income'], errors='coerce')
df['height'] = pd.to_numeric(df['height'], errors='coerce')

```

# Final cleanup

```python
df.drop_duplicates(inplace=True)  # Remove duplicates
df.reset_index(drop=True, inplace=True)  # Reset index
```

# Word Frequency Analysis (NLP on Essays)

```python
from collections import Counter
import re

# Combine all essays into one text corpus
text = ' '.join(df[['essay0', 'essay1', 'essay2', 'essay3', 'essay4', 'essay5', 'essay6', 'essay7', 'essay8', 'essay9']].fillna('').values.flatten())

# Remove punctuation and convert to lowercase
text = re.sub(r'[^a-zA-Z\s]', '', text.lower())

# Tokenize words and count frequency
words = text.split()
word_counts = Counter(words)

# Get the top 10 most common words
print(word_counts.most_common(10))
```

# Remove Stopwords

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Filter out stopwords
filtered_words = [word for word in words if word not in stop_words]
word_counts = Counter(filtered_words)

print(word_counts.most_common(10))
```

# Sentiment Analysis (Positive/Negative/Neutral)

```python
#pip install textblob
```

```python
from textblob import TextBlob

# Function to get sentiment polarity (-1 to 1)
def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity

# Apply to all essays
for i in range(10):  # essay0 to essay9
    df[f'essay{i}_sentiment'] = df[f'essay{i}'].apply(get_sentiment)

# Get overall sentiment statistics
df['overall_sentiment'] = df[[f'essay{i}_sentiment' for i in range(10)]].mean(axis=1)

# Categorize into Positive, Neutral, and Negative
df['sentiment_category'] = df['overall_sentiment'].apply(lambda x: 'Positive' if x > 0.1 else ('Negative' if x < -0.1 else 'Neutral'))

# Display sentiment distribution
print(df['sentiment_category'].value_counts())
```

# Topic Modeling (Extract Key Themes)

```python
#pip install sklearn
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Combine all essays into a single column
df['all_essays'] = df[[f'essay{i}' for i in range(10)]].fillna('').agg(' '.join, axis=1)

# Vectorize text using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X = vectorizer.fit_transform(df['all_essays'])

# Apply LDA to extract 5 topics
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(X)

# Print top words in each topic
words = vectorizer.get_feature_names_out()
for i, topic in enumerate(lda.components_):
    print(f"Topic {i+1}: ", [words[i] for i in topic.argsort()[-10:]])
```

Interpretation
Each topic consists of words that frequently appear together. This can help us find themes like:

Hobbies & Interests

Work & Career

Relationships & Dating

Travel & Adventure

Philosophy & Opinions

# User behaviour analysis 

Lifestyle Trends: Drinks, Drugs, and Smokes by Age Group
We will analyze how drinking, smoking, and drug habits vary across different age groups.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define age groups
bins = [18, 25, 35, 45, 55, 65, 100]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Count plot for lifestyle habits across age groups
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='age_group', hue='drinks', order=labels)
plt.title("Drinking Habits by Age Group")
plt.show()

plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='age_group', hue='smokes', order=labels)
plt.title("Smoking Habits by Age Group")
plt.show()

plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='age_group', hue='drugs', order=labels)
plt.title("Drug Use by Age Group")
plt.show()
```

Insights
Identify which age groups are more likely to drink, smoke, or use drugs.

See if younger or older people have healthier lifestyle choices.
