# Standardize Categorical Data

```python
# Standardizing categorical fields
categorical_columns = ['diet', 'body_type', 'education', 'drugs', 'smokes']

# Convert to lowercase and strip spaces
df[categorical_columns] = df[categorical_columns].apply(lambda x: x.str.lower().str.strip())

# Replace inconsistencies manually if needed
df['diet'].replace({'strictly any': 'strict', 'mostly oth': 'mostly other'}, inplace=True)
df['body_type'].replace({'a little extr': 'a little extra'}, inplace=True)
df['education'].replace({'working on': 'in progress', 'graduated from masters program': 'master’s degree'}, inplace=True)
df['drugs'].replace({'never': 'no', 'sometimes': 'occasional'}, inplace=True)
df['smokes'].replace({'sometimes': 'occasional', 'no': 'never'}, inplace=True)
```

# Fix Encoding Issues in Text Fields

```python
# List of essay columns
essay_cols = ['essay0', 'essay1', 'essay2', 'essay3', 'essay4', 'essay5', 'essay6', 'essay7', 'essay8', 'essay9']

# Fix encoding issues and remove HTML tags
import html

for col in essay_cols:
    df[col] = df[col].astype(str)  # Convert to string
    df[col] = df[col].apply(lambda x: html.unescape(x))  # Fix encoding issues
    df[col] = df[col].str.replace(r'<br\s*/?>', ' ', regex=True)  # Remove HTML tags
```

# Separate Multiple Categories

```python
# Split ethnicity into multiple binary columns
ethnicities = df['ethnicity'].str.get_dummies(sep=', ')
df = pd.concat([df, ethnicities], axis=1)

# Drop the original ethnicity column
df.drop(columns=['ethnicity'], inplace=True)
```

# Convert Numerical Fields to Proper Formats

```python

# Convert to numeric, forcing errors to NaN if necessary
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['income'] = pd.to_numeric(df['income'], errors='coerce')
df['height'] = pd.to_numeric(df['height'], errors='coerce')

```

# Final cleanup

```python
df.drop_duplicates(inplace=True)  # Remove duplicates
df.reset_index(drop=True, inplace=True)  # Reset index
```
